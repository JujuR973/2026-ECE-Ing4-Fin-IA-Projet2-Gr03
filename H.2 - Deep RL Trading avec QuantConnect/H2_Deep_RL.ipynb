```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
tickers = ["AAPL", "MSFT", "NVDA", "GOOGL", "AMZN"]

data = yf.download(
    tickers,
    start="2018-01-01",
    end="2023-01-01",
    interval="1d"
)


# Séparer prix et volume
price_df = data["Close"]
volume_df = data["Volume"]

print(price_df.head())
print(volume_df.head())
```
